```python
# dataframe init
df_stdout = pd.read_csv('./results_stdout.csv')
df_stderr = pd.read_csv('./results_stderr.csv')
df_kernel = pd.read_csv('./results_kernel.csv')

# dataframe columns update
df_stdout.columns = ["Group", "Type", *df_stdout.columns[2:]]
df_kernel.columns = ["Kernel", *df_kernel.columns[1:]]
```

```python
# stdout results charts
df_fp32 = df_stdout[df_stdout["Group"] == "fp32"]
df_gp = df_stdout[df_stdout["Group"] == "gp"]
df_fp32.index = df_fp32["Type"]
df_gp.index = df_gp["Type"]
df_fp32.plot(kind="bar", ylabel="error rate %", title="FP32 error measurements", rot=30)
df_gp.plot(kind="bar", ylabel="error rate %", title="GP error measurements", rot=30)
```
